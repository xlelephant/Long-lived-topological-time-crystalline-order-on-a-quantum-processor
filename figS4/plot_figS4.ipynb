```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
figS4_data = scipy.io.loadmat("figS4.mat", simplify_cells=True)
```

# FigS4.A, B

```python
ts_log10 = figS4_data["ts_log10"]
Zl1_avg = figS4_data["Zl1_avg"]
Zl1_err = figS4_data["Zl1_err"]
Ns = figS4_data["Ns"]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4), dpi=200)
for i in range(len(Ns)):
    axes[0].errorbar(
        ts_log10[i],
        Zl1_avg[i],
        Zl1_err[i],
        label=f"N={Ns[i]}",
        elinewidth=1.3,
        markersize=3,
        marker="o",
        ls="-",
        linewidth=1.5,
        capsize=3.0,
        capthick=0.5,
    )
axes[0].plot([0, np.max(ts_log10)], [0.5] * 2, "-.k", label="0.5")
axes[0].legend()
axes[0].grid(1)

axes[0].set_ylabel(r"$|{\it{Z_{L_1}}}|$")
axes[0].set_xticks(
    [0, 3, 6],
    [r"$10^" + f"{Ni}" + r"$" for Ni in [0, 3, 6]],
)
axes[0].set_xlabel("t / T")

T_th_log10 = figS4_data["T_th_log10"]
T_th_log10_fit = figS4_data["T_th_log10_fit"]
axes[1].plot(Ns, T_th_log10_fit, "-.k")
for i in range(len(Ns)):
    axes[1].plot(Ns[i], T_th_log10[i], markersize=6, marker="o")
axes[1].grid(1)
axes[1].set_yticks(
    [3, 5, 7],
    [r"$10^" + f"{Ni}" + r"$" for Ni in [3, 5, 7]],
)
axes[1].set_ylim([2.2, 7.2])
axes[1].set_ylabel(r"$\it{\tau*/T}$")
axes[1].set_xticks(Ns)
axes[1].set_xlabel("System size")
```
