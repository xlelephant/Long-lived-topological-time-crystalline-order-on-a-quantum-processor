```python
import numpy as np
import matplotlib.pyplot as plt
import scipy


def ecdf(data: list, bins=None):
    bins_ = 1
    if bins is None:
        bins = np.sort([min(data), max(data)] + data)
        bins_ = None
    his_vals, base = np.histogram(data, bins=bins)
    cumulative = np.cumsum(his_vals)
    if bins_ is None:
        return base, cumulative / max(cumulative)
    else:
        return base, (cumulative - 1) / (len(data) - 1)
```

```python
figS7_data = scipy.io.loadmat("figS7.mat", simplify_cells=True)
```

```python
f10_list = list(figS7_data["f10"].values())
base, steps = ecdf(f10_list)
plt.step(base[1:], steps, color="#244c9c")
plt.plot([np.median(f10_list)] * 2, [0, 1], "--", color="#244c9c")
plt.grid()
plt.ylim([0, 1.0])
plt.ylabel("Integrated histogram")
plt.xlabel("Freuquency (GHz)")
plt.title(
    r"Qubit $\frac{\omega_{10}}{2\pi}$"
    + "\n Median: {:.1f} GHz".format(np.median(f10_list))
)
```

```python
T1_list = list(figS7_data["T1"].values())
base, steps = ecdf(T1_list)
plt.step(base[1:], steps, color="#244c9c")
plt.plot([np.median(T1_list)] * 2, [0, 1], "--", color="#244c9c")
plt.grid()
plt.ylim([0, 1.0])
plt.xlim([0, 220])
plt.ylabel("Integrated histogram")
plt.xlabel(r"Delay ($\mu$s)")
plt.title(r"Qubit $T_1$" + "\n Median: {:.1f}".format(np.median(T1_list)) + r"$ \mu$s")
```
