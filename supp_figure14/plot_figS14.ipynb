```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps
import scipy
import matplotlib

label_colors_1 = list(matplotlib.colors.TABLEAU_COLORS.keys())

label_dict_Z = {
    "0": r"$Z_{L_1}$",
    "1": r"$Z_{L_2}$",
    "2": r"$Z_{L_3}$",
    "3": r"$Z_{L_4}$",
    "4": r"$Z_{L_5}$",
    "5": r"$Z_{L_6}$",
}

fig = plt.figure(figsize=(8.27 * 0.85, 8.27 * 8 / 12 * 0.85), dpi=300)
axes = fig.subplots(nrows=len(label_dict_Z), ncols=1, sharex=True)

rst_echo = scipy.io.loadmat('figS14.mat', simplify_cells=True)
for i in range(len(label_dict_Z)):
    ax = axes[i]
    line = ax.plot(
        rst_echo["Ts"],
        rst_echo["Zs_echo_avg"][i, :],
        color="grey",
    )
    ax.errorbar(
        rst_echo["Ts"],
        rst_echo["Zs_exp_avg"][i, :],
        rst_echo["Zs_exp_err"][i, :],
        color=label_colors_1[i],
        elinewidth=1.0,
        ls="none",
        linewidth=0.8,
        capsize=1.8,
        capthick=1.2,
    )
    ax.plot(
        rst_echo["Ts"],
        rst_echo["Zs_sim_avg"][i, :],
        color=label_colors_1[i],
        alpha=0.7,
    )
    ax.grid()
    ax.set_xlim([0, 20])
    ax.set_ylabel("$\overline{A^{1/3}_{L_%d}}(t)$" % (i + 1))
ax.set_xticks(np.arange(0, 21, 2))
ax.set_xticklabels(np.arange(0, 21, 2))
ax.set_xlabel("t/T")
```
