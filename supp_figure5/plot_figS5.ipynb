```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io

rst_exp = scipy.io.loadmat(f"./figS5.mat", simplify_cells=True)
fig = plt.figure(figsize=(8.27 * 0.85 * 0.6, 8.27 * 8 / 12 * 0.85 * 0.7), dpi=300)
for j, disorder in enumerate(rst_exp['disorders']):
    plt.errorbar(
        rst_exp["Ts"],
        rst_exp["Fid_cir_step_avg"][j],
        rst_exp["Fid_cir_step_sem"][j],
        marker="o",
        color=["#365ba3", "#e89c40"][j],
        elinewidth=1.3,
        linewidth=0.8,
        capsize=1.8,
        capthick=1.2,
        label=disorder,
    )
plt.xlabel(r"$t/T$")
plt.ylabel(r"$F(\rho)$")
plt.title("Four-qubit subsystem state fidelity")
plt.grid(True)
plt.legend()
plt.tight_layout()
```
